```python
#creating histograms for audio features
import matplotlib.pyplot as plt
#a list col_names created
col_names = ['Sale Price','Mrp','Star Rating','Discount Percentage','Number Of Ratings','Number Of Reviews']

#new dataframe audi_features is created with columns specified in col_names
features = df[col_names]

#figure with size 15*12 is created for multiplot
fig = plt.figure(figsize=(15, 12))

#grid of subplots with 4rows and 3 columns, with a spacing of 0.5 between rows and columns is created
gs = fig.add_gridspec(2,3 , hspace=0.5, wspace=0.5)

#an array of subplots is created is created based on the grid
axs = gs.subplots()

#title
fig.suptitle("Histograms")

#loop to iterate over each subplot
for i, ax in enumerate(axs.flatten()):

  #values of corresponding audio features is accessed, i is the index of subplot
    vals = features.iloc[:, i].values
    title = features.columns[i]

    #histogram created
    ax.hist(vals)

    #title of each subplot is set
    ax.set_title(f'{title}')

#to display the figure
plt.show()
```

```python
df = df.dropna()
df = df.drop_duplicates()
top10 = df.sort_values(by='Star Rating', ascending=False).head(10)
top10
```

```python
import numpy as np
#Difference between the mrp and sale price of the highest rated products
plt.figure(figsize=(16,6))
X_axis = np.arange(len(top10['Product Name']))
plt.bar(X_axis-0.2,top10['Mrp'],0.4,label='MRP')
plt.bar(X_axis+0.2,top10['Sale Price'],0.4,label='Sale Price')
plt.xticks(X_axis, top10['Product Name'],rotation=90)
plt.legend()
plt.show()
```

```python
low10 = df.sort_values(by='Star Rating', ascending=True).head(10)
low10

#Difference between the mrp and sale price of the lowest rated products
plt.figure(figsize=(16,6))
X_axis = np.arange(len(low10['Product Name']))
plt.bar(X_axis-0.2,low10['Mrp'],0.4,label='MRP')
plt.bar(X_axis+0.2,low10['Sale Price'],0.4,label='Sale Price')
plt.xticks(X_axis, low10['Product Name'],rotation=90)
plt.legend()
plt.show()
```
